```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


sns.set(font_scale=1.9)
sns.set_style("white")


x1 = [np.random.randint(10,60) for i in range(1000)]
x2 = [np.random.randint(10,60) for i in range(1000)]
x3 = [np.random.randint(10,60) for i in range(1000)]
x4 = [np.random.randint(40,60) for i in range(1000)]

kwargs = dict(hist_kws={'alpha':.8}, kde_kws={'linewidth':3})
 
fig, axes = plt.subplots(2, 2, figsize=(19, 15))
 
fig.suptitle('Disttribution Charts')
 
 
sns.distplot(a=x1, ax=axes[0, 0], color="dodgerblue", label="Compact", **kwargs )
sns.distplot(a=x3, ax=axes[0, 1], color="deeppink", label="minivan", **kwargs)
sns.distplot(a=x2, ax=axes[1, 0], color="orange", label="SUV", **kwargs)
sns.distplot(a=x4, ax=axes[1, 1], color="red", label="SUZ", **kwargs)


fig.tight_layout() 
plt.savefig('distchart.png')

plt.show()

             
```
