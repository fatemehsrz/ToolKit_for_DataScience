```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


sns.set(font_scale=1.9)
sns.set_style("white")

# Import data
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/diamonds.csv')
x1 = df.loc[df.cut=='Ideal', 'depth']
x2 = df.loc[df.cut=='Fair', 'depth']
x3 = df.loc[df.cut=='Good', 'depth']
x4 = [np.random.randint(40,80) for i in range(10000)]
 
fig, axes = plt.subplots(2, 2, figsize=(19, 15))
 
fig.suptitle('Disttribution Charts')
 
 
sns.distplot(a=x1, ax=axes[0, 0], color="dodgerblue", label="Compact")
sns.distplot(a=x3, ax=axes[0, 1], color="deeppink", label="minivan")
sns.distplot(a=x2, ax=axes[1, 0], color="orange", label="SUV")
sns.distplot(a=x4, ax=axes[1, 1], color="red", label="SUZ")

fig.tight_layout() 
plt.savefig('dischart.png')
plt.show()
             
```
